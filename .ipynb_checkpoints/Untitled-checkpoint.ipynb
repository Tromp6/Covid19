```python
def formateDate(date):
    dateList = date.split()[0].split("/")
    year = [dateList.pop()]
    
    for i,x in enumerate(dateList):
        if len(x) == 1:
            dateList[i] = "0"+x
    dateList = year + dateList
    dateInt = int(''.join(dateList))
    return(dateInt)

def formateToDateObject(date):
    return pd.to_datetime(date).date()
```

```python
data['formated_ref_date'] = data['Referenzdatum'].apply(lambda date: formateToDateObject(date))
```

```python
temp = data.loc[data["AnzahlFall"] > 0].sort_values(by=['formated_ref_date']).groupby(['formated_ref_date','Altersgruppe'])['AnzahlFall'].sum()
```

```python
dictData = {'Referenzdatum':[],'Altersgruppe':[],'AnzahlFall':[]}
for i,x in enumerate(temp):
    #comprassedDate = '/'.join(temp.index[i][0].split('/')[:2])
    dictData['Referenzdatum'].append(temp.index[i][0])
    dictData['Altersgruppe'].append(temp.index[i][1])
    dictData['AnzahlFall'].append(x)
```

```python
fig, ax = plt.subplots(figsize=(25,10))
plt.xticks(rotation=45)
sns.lineplot(ax=ax, x='Referenzdatum', y='AnzahlFall', hue='Altersgruppe', 
             data=dictData,sort=False)
```
